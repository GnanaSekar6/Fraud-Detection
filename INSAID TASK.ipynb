```python
data=data.drop(['nameOrig','nameDest'],axis=1) #removing these two columns since they doesn't seem helpful for the accuracy of the prediction
```

```python
# importing label encoder for converting each categorical value in to label
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
#splitting the data set into training and testing.

data['type']= label_encoder.fit_transform(data['type'])
X, y = data.loc[:, data.columns != 'isFraud'], data['isFraud']
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.40,random_state=42)
```

```python
#scaling the dataset to fit in to training and testing 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# MODEL

```python
#Import Gaussian Naive Bayes model

from sklearn.naive_bayes import GaussianNB


#Import scikit-learn metrics module for accuracy calculation

from sklearn import metrics 


#Create a Gaussian Classifier

gnb = GaussianNB()
```

```python
#Training the model using the training data sets

gnb.fit(X_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = gnb.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Prediction of accuracy using Logistic Regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,precision_score,recall_score
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,y_pred)
print("Accuracy:",result2)
print(precision_score(y_test,y_pred))
print(recall_score(y_test,y_pred))
```
